```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('Iris.csv')
df
```

```python
df.head()
```

```python
df.tail()
```

```python
df.sample(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(columns='Id', inplace=True)
df.head()
```

```python
dummies = pd.get_dummies(df.Species)
dummies = dummies.astype(int)
dummies.tail()
```

```python
merged_data = pd.concat([df,dummies], axis=1)
merged_data.head()
```

```python
final_data = merged_data.drop(columns='Species')
final_data.head()
```

```python
final_data.isnull().sum().sum()
```

```python
final_data.hist(figsize=(10,10))
plt.show()
```

```python
corr_mat = final_data.corr()
sb.heatmap(corr_mat, annot=True,cmap='coolwarm')
plt.show()
```

```python
plt.scatter(final_data['SepalLengthCm'],final_data['SepalWidthCm'],alpha=0.4,s=100*final_data['PetalLengthCm'],cmap='viridis')
```

```python
target_name= ['Iris-setosa','Iris-versicolor','Iris-virginica']

# Separate object for target feature
y = final_data[target_name]
y

# Separate object for input feature
x = final_data.drop(target_name, axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=1)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
from sklearn. ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
```

```python
print("Train Accuracy of Random Forest Algorithm ", rf.score(x_train,y_train)*100)
print("Test Accuracy score of Random Forest Algorithm ", rf.score(x_test,y_test)*100)
```

```python
prediction = rf.predict([[6.3,4.1,2,0.9]])
print(prediction)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
```

```python
print("Train Accuracy of Decision Tree Algorithm ", dt.score(x_train,y_train)*100)
print("Test Accuracy score of Decision Tree Algorithm ", dt.score(x_test,y_test)*100)
```

```python
prediction = dt.predict([[6.3,4.1,2,0.9]])
print(prediction)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

```python
print("Train Accuracy of KNN Algorithm ", knn.score(x_train,y_train)*100)
print("Test Accuracy score of KNN Algorithm ", knn.score(x_test,y_test)*100)
```

```python
prediction = knn.predict([[6.3,4.1,2,0.9]])
print(prediction)
```
